```python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
pl=Person("aman",20)
print(pl.name)
print(pl.age)
```

```python
class circle():
  def __init__(self,r):
    self.radius=r
  def area(self):
    return self.radius**2*3.14
  def perimeter(self):
    return 3.14*2*self.radius
new_circle=circle(7)
print("circle's area is:",new_circle.area())
print("circle's perimeter is:",new_circle.perimeter())
```

```python
class circle:
  def __init__(self,r):
    self.radius=r
  def area(self):
    return 3.14*self.radius**2
  def perimeter(self):
    return 2*3.14*self.radius
new_circle=circle(78)
print("area of the new circle is:",new_circle.area())
print("perimeter of the new circle: ",new_circle.perimeter())
```
