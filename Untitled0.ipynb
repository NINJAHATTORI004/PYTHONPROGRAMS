```python
# prompt:  Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales

import matplotlib.pyplot as plt

# Group the data by month and vehicle type
grouped_data = df.groupby(['month', 'vehicle_type'])['sales'].sum().unstack()

# Create the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(grouped_data.index, grouped_data.columns, s=grouped_data.values * 2, alpha=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Vehicle Type')
plt.title('Impact of Seasonality on Automobile Sales')

# Add colorbar
plt.colorbar()

# Show the plot
plt.show()

# Analyze the trends:
# - Look for patterns in sales volume across different months for each vehicle type.
# - Identify any vehicle types that show significant seasonal variations in sales.
# - Draw conclusions about the impact of seasonality on automobile sales based on the observed trends.

# Example conclusion:
# The bubble plot reveals that sales of convertibles and motorcycles peak during the summer months, indicating a strong influence of seasonality on these vehicle types. In contrast, sales of SUVs and trucks remain relatively stable throughout the year, suggesting a lesser impact of seasonality.
```

```python
# prompt:  Create and display graphs for Yearly Report Statistics

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create some example data
data = {
    'Year': [2018, 2019, 2020, 2021],
    'Sales': [10000, 12000, 8000, 15000],
    'Expenses': [5000, 6000, 4000, 7000],
    'Profit': [5000, 6000, 4000, 8000]
}

df = pd.DataFrame(data)

# Create a line chart for sales, expenses, and profit
plt.figure(figsize=(15, 5))
plt.plot(df['Year'], df['Sales'], label='Sales')
plt.plot(df['Year'], df['Expenses'], label='Expenses')
plt.plot(df['Year'], df['Profit'], label='Profit')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Yearly Sales, Expenses, and Profit Trends')
plt.legend()
plt.show()

# Create a bar chart for sales by year
plt.figure(figsize=(10, 5))
plt.bar(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales Comparison')
plt.show()

# Create a pie chart for profit distribution
plt.figure(figsize=(8, 8))
plt.pie(df['Profit'], labels=df['Year'], autopct='%1.1f%%')
plt.title('Profit Distribution by Year')
plt.show()
```
